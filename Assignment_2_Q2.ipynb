```python
def keyRelBasedList(df):
  temp_list = []
  for j in range(len(df)):
    temp_list.append((list(df.index)[j],list(df[0])[j]))

  return temp_list
```

```python
def maxDcg(tmp_list):
  max_dcg = 0;
  for j in range(len(tmp_list)):
    if j == 0:
            max_dcg = tmp_list[j][1]
    else:
            max_dcg += tmp_list[j][1]/math.log2(j+1) 
  return max_dcg          
```

```python
max_dcg = maxDcg(idxRelList)
print("Max DCG for ideal case is: ",max_dcg)
```

```python
def totalfiles(tmp_list):
   count_list = [0] * 4
   for i in range(len(tmp_list)):
     if tmp_list[i][1] == 0:
       count_list[0] +=1
     elif tmp_list[i][1] == 1:
       count_list[1] +=1
     elif tmp_list[i][1] == 2:
       count_list[2] +=1
     elif tmp_list[i][1] == 3:
       count_list[3] +=1    
   print(str(count_list[3]) + "! * " + str(count_list[2]) + "! * " + str(count_list[1]) + "! * " + str(count_list[0]) + "!")   
   return math.factorial(count_list[3])*math.factorial(count_list[2])*math.factorial(count_list[1])*math.factorial(count_list[0])
```

```python
print("Total file with max dcg can be:  ", totalfiles(idxRelList))
```

```python
def nDCG(df,stoppingIdx):
    tmp_list = []
    idx_list =  list(df.index)
    rel_list = list(df[0])
    for j in range(len(idx_list)):
        tmp_list.append((idx_list[j],rel_list[j]))

    dcg = 0 
    for j in range(len(tmp_list)):
      if j == 0:
              dcg = tmp_list[j][1]
      else:
              dcg += tmp_list[j][1]/math.log2(j+1) 
      if( j == stoppingIdx ):
          break
    dcg = dcg/max_dcg
    
    return dcg
```

```python
ndcg_all = nDCG(df_txt_q4,-100)
ndcg_50 = nDCG(df_txt_q4,50)
print("NDCG for 50",ndcg_50)
print("NDCG for all",ndcg_all)
```

```python

def relIdf(df):
  tmp_list = []
  rel_list = list(df[0])
  idf_list = (list(df[76]))
  for j in range(len(rel_list)):
    tmp_list.append((rel_list[j], float(idf_list[j].split(':')[1])))
  tmp_list = sorted(tmp_list,key=lambda t: t[1],reverse =True)
  return tmp_list 

def findNumRelDocs(tmp_list):
  num_rel_docs = 0
  for j in range(len(tmp_list)):
      if tmp_list[j][0] > 0:
          num_rel_docs+=1
  return num_rel_docs        

def draw_graph(prec_list,rec_list):
    plt.plot(rec_list,prec_list)
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title('Precision V/S Recall Curve')
    plt.show() 

def calculatePrecRec(df):
    tmp_list = relIdf(df)
    num_rel_docs = findNumRelDocs(tmp_list)
    rel_curr_docs = 0
    val=0
    prec_list = []
    rec_list = []
    for i in range(len(tmp_list)):
        val+=1
        if tmp_list[i][0] > 0:
            rel_curr_docs+=1
        prec_list.append(rel_curr_docs/(val))
        rec_list.append(rel_curr_docs/num_rel_docs)
    draw_graph(prec_list,rec_list)
```

```python
calculatePrecRec(df_txt_q4)
```
