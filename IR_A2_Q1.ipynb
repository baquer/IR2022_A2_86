```python
def lower_case(text):
    lower_case_text = text.lower()
    return lower_case_text
```

```python
def tokenize(text):
  tokenizer = TweetTokenizer()
  token_list = tokenizer.tokenize(text)
  return token_list
```

```python
def stop_word(text):
  sentence = []
  stop_words = set(stopwords.words("english"))
  for w in text:
        if w not in stop_words:
            sentence.append(w)
        else:
          continue
  return sentence
```

```python
def remove_punc(text):
  punc_tokenizer = nltk.RegexpTokenizer(r"\w+")
  text = punc_tokenizer.tokenize(text)
  return text
```

```python
def tokenizeReg(text):
    tok=RegexpTokenizer('[A-Za-z0-9]?\w+')
    return tok.tokenize(text)
```

```python
def lemmatization(file):
  lemmatizer = WordNetLemmatizer()
  for i in range(0,len(file)):
    lemma = lemmatizer.lemmatize(file[i])
    file[i] = lemma
  return file  
```

```python
def jaccardCofficient(query):
  content = ""
  jaccardDict = {}
  for filename in glob.glob(os.path.join(path+"/", '*')):
    f=open(filename,'r',errors='ignore')
    content=f.read()
    content = lower_case(content)
    content=tokenizeReg(content)
    content = [word for word in content if word.isalpha()]
    new_data = [word for word in content if not word.isdigit()]
    new_data=lemmatization(new_data)
    new_data=stop_word(new_data)
    content = set(new_data)
    query_content = lower_case(query)
    query_content=tokenizeReg(query_content)
    query_content = [word for word in query_content if word.isalpha()]
    new_data1 = [word for word in query_content if not word.isdigit()]
    new_data1 = lemmatization(new_data1)
    new_data1 = stop_word(new_data1)
    query_content=set(new_data1)

    inter = set.intersection(content, query_content)

    uni = set.union(content, query)
  

    jaccard = len(inter)/len(uni)

    if jaccard != 0.0:
      jaccardDict[filename] = jaccard
      #print(inter)
  k = Counter(jaccardDict)
  getTopFive = k.most_common(5) 
  print("Document :    Jaccard Cofficient")  
  for i in getTopFive:
    print("{}    :{}".format(i[0],i[1]))



```

```python
# Document Frequency 
# Compute item count
def documentFreq(finalDict, vocab):
  dict3 = {}
  for word in vocab:
    count = 0
    for k in finalDict.keys():
      if word in finalDict[k].keys():
        count += 1
    dict3[word] = count
  return dict3
```

```python
import math
def binary(df,dict1,vocab,noOfDocs):
  #Matrix of no of docs vs vocab
  mat = np.zeros(shape=(noOfDocs,len(vocab)))
  data = pd.DataFrame(mat,columns=vocab)
  for i in range(0, noOfDocs):
    for term in dict1[i].keys():
      # compute IDF
      idf = math.log(noOfDocs/(df[term]+1))
      #if idf != 0.0:
        #print(idf)
      data[term][i]=idf
  return data
```

```python
def raw(df,dict1,vocab,noOfDocs):
    mat = np.zeros(shape=(noOfDocs,len(vocab)))
    data = pd.DataFrame(mat,columns=vocab)
    for i in range(0, noOfDocs):
       for term in dict1[i].keys():
            idf = math.log(noOfDocs/(df[term]+1))
            tf = dict1[i][term]
            data[term][i] = tf * idf
    return data
```

```python
#Size of corpus of each document
corpusLen = {}

for i in globalContent.keys():
  corpusLen[i] = len(globalContent[i])
```

```python
def termF(df,dict1,vocab,noOfDocs):
    mat = np.zeros(shape=(noOfDocs,len(vocab)))
    data = pd.DataFrame(mat,columns=vocab)
    for i in range(0, noOfDocs):
       for term in dict1[i].keys():
            idf = math.log(noOfDocs/(df[term]+1))
            tf = dict1[i][term]/corpusLen[i]
            data[term][i] = tf * idf
    return data
```

```python
def log(df,dict1,vocab,noOfDocs):
    mat = np.zeros(shape=(noOfDocs,len(vocab)))
    data = pd.DataFrame(mat,columns=vocab)
    for i in range(0, noOfDocs):
       for term in dict1[i].keys():
            idf = math.log(noOfDocs/(df[term]+1))
            tf = math.log(dict1[i][term] + 1)
            data[term][i] = tf * idf
    return data
```

```python
# max in each doc

maxLen = {}

maxx = 0

for i in globalContent.keys():
  for value in globalContent[i]:
    maxLen[i] = max(maxx, len(value))

```

```python
def doubleT(df,dict1,vocab,noOfDocs):
    mat = np.zeros(shape=(noOfDocs,len(vocab)))
    data = pd.DataFrame(mat,columns=vocab)
    for i in range(0, noOfDocs):
       for term in dict1[i].keys():
            idf = math.log(noOfDocs/(df[term]+1))
            tf = 0.5 + 0.5*(dict1[i][term]/maxLen[i])
            data[term][i] = tf * idf
    return data
```
